```python
"""
Alexis VENET, CPE Lyon, 4ETI
9/04/25
"""
# Classe Noeud et ListeChainee (inchangées sauf pour correction de petits bugs)
class Noeud:
    def __init__(self, valeur):
        self.next = None
        self.val = valeur
    #permet d'avoir le prochain noeud et la valeur de celui ci


    def __repr__(self):
        return f"{self.val}"

class ListeChainee:
    def __init__(self):
        self.tete = None

    def init_debut(self, valeur):
        nouveau = Noeud(valeur)
        nouveau.next = self.tete
        self.tete = nouveau
    #permet de définir un début de listechainé, avec un noeud de tete

    def ajouter_fin(self, valeur):
        nouveau = Noeud(valeur)
        if not self.tete:
            self.tete = nouveau
        else:
            courant = self.tete
            while courant.next:
                courant = courant.next
            courant.next = nouveau
    #permet d'ajouter un noeud à la fin
    
    def __repr__(self):
        elements = []
        courant = self.tete
        while courant:
            elements.append(repr(courant))
            courant = courant.next
        return " -> ".join(elements)
    #permet de réaliser de l'affichage
    
    def inserer(self, valeur, k):
        nouveau = Noeud(valeur)
        if k == 0:
            nouveau.next = self.tete
            self.tete = nouveau
            return
        courant = self.tete
        i = 0
        while courant.next and i < k - 1:
            courant = courant.next
            i += 1
        nouveau.next = courant.next
        courant.next = nouveau
    #permet de mettre un noeud à l'indice souuhaité

    def supprimer(self, k):
        if k == 0:
            self.tete = self.tete.next
            return
        courant = self.tete
        i = 0
        while courant.next and i < k - 1:
            courant = courant.next
            i += 1
        if courant.next:
            courant.next = courant.next.next
    #permet de supprimer un noeud à l'indice souhaité

    def rechercher(self, val):
        courant = self.tete
        while courant:
            if courant.val == val:
                print(f"{val} : valeur présente")
                return True
            courant = courant.next
        print(f"{val} : valeur non trouvée")
        return False
    #permet de savoir si la valeur est présente dans la liste chainée

    def obtenir(self, k):
        i = 0
        courant = self.tete
        while courant and i < k:
            courant = courant.next
            i += 1
        if courant:
            print(f"val à l'indice {k} : {courant.val}")
            return courant.val
        return None
    #permet d'obtenir la valeur du noeud à l'indice k

    def taille(self):
        i = 0
        courant = self.tete
        while courant:
            i += 1
            courant = courant.next
        return i
    #retourne la taille de la liste chainé

    def estVide(self):
        return self.tete is None
    #Renvoie True si la tete est vide, False sinon

    def supprimer_valeur(self, val):
        courant = self.tete
        precedent = None
        while courant:
            if courant.val == val:
                if precedent:
                    precedent.next = courant.next
                else:
                    self.tete = courant.next
                return True
            precedent = courant
            courant = courant.next
        return False
    #permet de supprimer un noeud avec une valeur

# Exercice 1 - Question 3 & 4 : Train avec montée/descente
class Train:
    def __init__(self, nb_wagons):
        self.wagons = [ListeChainee() for _ in range(nb_wagons)]
    #crée nb_wagon liste chainé

    def afficher(self):
        print("\nTrain :")
        for i, wagon in enumerate(self.wagons):
            print(f"Wagon {i}: {wagon}")
        print()
    #fonction d'affichage pour le train

    def monter(self, passager, k):
        print(f"Passager {passager} essaie de monter dans le wagon {k}")
        # Tentative vers l'avant du train si fin pleine
        for i in range(k, len(self.wagons)):
            if self.wagons[i].taille() < 4:
                self.wagons[i].ajouter_fin(passager)
                self.afficher()
                return
    
        # Revenir en arrière
        for i in range(k - 1, -1, -1):
            if self.wagons[i].taille() < 4:
                self.wagons[i].ajouter_fin(passager)
                self.afficher()
                return
        print("Le train est plein. Le passager ne peut pas monter.")
        self.afficher()
    #permet à un passager de monter dans le train
    #l'empeche de monter si plus de place


    def descendre(self, passager):
        for i, wagon in enumerate(self.wagons):
            if wagon.supprimer_valeur(passager):
                print(f"Passager {passager} est descendu du wagon {i}")
                self.afficher()
                return
        print(f"Passager {passager} n'était pas dans le train.")
        self.afficher()
    #permet de descendre un passager si présent dans le train

# Exercice 2 - Question 1 : Pile
class Pile(ListeChainee):
    def empiler(self, val):
        self.init_debut(val)

    def depiler(self):
        if self.estVide():
            return None
        val = self.tete.val
        self.tete = self.tete.next
        return val

    def sommet(self):
        if self.estVide():
            return None
        return self.tete.val

# Exercice 2 - Question 2 : Vérification des parenthèses
def parenthese_sentence(expr):
    correspondance = {')': '(', ']': '[', '}': '{', '>': '<'}
    pile = Pile()
    for char in expr:
        if char in "({[<":
            pile.empiler(char)
        elif char in ")}]>":
            if pile.estVide() or pile.depiler() != correspondance[char]:
                return False
    return pile.estVide()
```

### Affichage noeud et liste chainée

```python
liste = ListeChainee()
liste.init_debut(3)
liste.init_debut(2)
liste.init_debut(1)

liste.inserer(10,1)

liste.ajouter_fin(4)
liste.ajouter_fin(5)
print(f"avant suppression : {liste}")
liste.supprimer(1)
print(f"après suppression : {liste}")
liste.rechercher(5)
liste.obtenir(1)
print(f"taille : {liste.taille()}")
print(liste)
```
