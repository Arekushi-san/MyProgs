```python
import numpy as np
import matplotlib.pyplot as plt

def fct1(t, y1, y2, y3):
    return y2

def fct2(t, y1, y2, y3):
    return y3

def fct3(t, y1, y2, y3):
    return t * np.exp(-t**2) - 1/y1

def fct_Euler(y1_0, y2_0, y3_0, t_min, t_max, step):
    """
    Implémente la méthode d'Euler pour résoudre une équation différentielle y'(t) = f(t, y).

    Arguments :
    - y0 : condition initiale (valeur de y à t = tmin)
    - tmin : début de l'intervalle de temps
    - tmax : fin de l'intervalle de temps
    - step : pas de discrétisation
    - fonction : fonction f(t, y) définissant l'équation différentielle

    Retourne :
    - y : tableau des valeurs de y(t)
    - t : tableau des instants de temps
    """
    t_values = np.arange(t_min, t_max + step, step)
    y1_values = np.zeros(len(t_values))
    y2_values = np.zeros(len(t_values))
    y3_values = np.zeros(len(t_values))
    
    # Conditions initiales
    y1_values[0] = y1_0
    y2_values[0] = y2_0
    y3_values[0] = y3_0
    
    for k in range(len(t_values) - 1):
        t = t_values[k]
        y1, y2, y3 = y1_values[k], y2_values[k], y3_values[k]
        
        y1_values[k+1] = y1 + step * fct1(t, y1, y2, y3)
        y2_values[k+1] = y2 + step * fct2(t, y1, y2, y3)
        y3_values[k+1] = y3 + step * fct3(t, y1, y2, y3)
    
    return t_values, y1_values




# Conditions initiales
y1_0 = 1
y2_0 = 0
y3_0 = 1
t_min = 0
t_max = 8
step = 0.1

# Résolution avec Euler
t_values, y1_values = fct_Euler(y1_0, y2_0, y3_0, t_min, t_max, step)

# Affichage des résultats
plt.plot(t_values, y1_values, label="Approximation par Euler")
plt.xlabel("t")
plt.ylabel("y(t)")
plt.legend()
plt.grid()
plt.show()

```
